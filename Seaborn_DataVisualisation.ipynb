```python
#RELPLOT
#1#Use relplot when there are two NUMERICAL attributes, or
#2#to find relation b/w two numerical and one categorical attributes

sns.relplot(x="Survived",y="Parch",color='g',data=df)
```

```python
#now put a categorical attribute as (hue)
sns.relplot(x="Survived",y="Parch",hue='Sex',data=df)
```

```python
#CATPLOT
#on X axis use Categorical attribute and on Y axis:Numerical attribute

sns.catplot(x="Embarked",y="Age",data=df)
#Embarked means the port from where the passenger got into the ship
```

```python
sns.catplot(x="Pclass",y="Age",data=df)
#INference from the data: Majority of old people were in class 1
```

```python
#using jitter
#All points come into same line(gives more clarity)
sns.catplot(x="Pclass",y="Age",jitter=False,data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",data=df)
#Infernce; no. of males are more in class 3
```

```python
sns.catplot(x="Pclass",y="Age",kind="swarm",data=df)
#gives a kind(i guess some sort of pattern) to the graph
```

```python
sns.catplot(x="Pclass",y="Age",kind="box",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",kind="violin",data=df)
```

```python
sns.catplot(x="Pclass",y="Survived",kind="violin",data=df)
```

```python
sns.catplot(x="Pclass",y="Survived",kind="violin",split=True,data=df)
#Makes the graph more FINER
```

```python
sns.catplot(x="Pclass",y="Age",kind="point",data=df)
```
